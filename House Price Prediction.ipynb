```python
import pandas as pd
```

```python
df=pd.read_csv("Multi.csv")
```

```python
df
```

```python
df.drop("price",axis=1)
```

```python
x=df[["area","bedrooms","age"]].values
y=df["price"].values
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression
import warnings
warnings.filterwarnings("ignore")
```

```python
m1=LinearRegression()
```

```python
m1.fit(x,y)
```

```python
m1.coef_
```

```python
m1.intercept_
```

```python
area=int(input("Enter the Area:"))
bedrooms=int(input("Enter the No. of Bedrooms:"))
age=int(input("Enter How Old House You Looking For:"))
pred=m1.predict([[area,bedrooms,age]])
print("The Price of Area is ",pred[0])
```
